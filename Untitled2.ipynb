Objective:

To analyze and visualize pharmaceutical data to identify the most commonly treated medical conditions and the distribution of drug classes, providing insights into healthcare priorities and treatment trends.

```python
data =data.dropna(axis =1, how ='all')
```

```python
columns_to_fill = ['side_effects', 'generic_name', 'drug_name', 'drug_classes']

for col in columns_to_fill:
    if data[col].isnull().sum() > 0:
        mode_value = data[col].mode()[0]
        data[col].fillna(mode_value, inplace=True)
```

```python
data.isnull().sum()
```

```python
data =data.dropna(axis =1, how ='any')
```

```python
sns.countplot(y ='medical_condition', data =data , order=data['medical_condition'].value_counts().index[:10])
plt.title('top 10 medical conditions treated')
plt.show()
```

 Key Observations:

  1.	Pain is the most commonly treated medical condition in the dataset, with over 250 drug entries, highlighting its widespread nature and the variety of treatments available.

  2.	Colds & Flu and Acne follow closely, reflecting the common and recurring nature of these conditions, especially among general and dermatological care.

  3.	Chronic conditions such as Hypertension, Osteoarthritis, and Diabetes (Type 2) also appear prominently, indicating a strong pharmaceutical focus on long-term disease management.

  4.	Conditions like AIDS/HIV, Eczema, and Psoriasis are also in the top 10, suggesting significant medical and research attention in immune-related and skin conditions.

  5.	The plot also shows that while some conditions have a large variety of treatment options (like pain), others may have fewer, possibly due to complexity in treatment or limited drug development.

```python
top_classes = data['drug_classes'].value_counts().head(10)
plt.figure(figsize=(7,7))
plt.pie(top_classes, labels=top_classes.index, autopct='%1.1f%%', startangle=140)
plt.title("Top 10 Drug Classes (Pie Chart)")
plt.axis('equal')
plt.show()
```

Insight:   The dominance of upper respiratory and topical treatments suggests a strong pharmaceutical focus on widely experienced, high-frequency conditions.

```python

plt.figure(figsize=(10, 6))
top_conditions = data['medical_condition'].value_counts().head(10)
top_conditions_df = top_conditions.reset_index()
top_conditions_df.columns = ['medical_condition', 'count']
sns.barplot(x='count', y='medical_condition', data=top_conditions_df)
plt.title("Top 10 Medical Conditions")
plt.xlabel("Number of Drugs")
plt.ylabel("Medical Condition")
plt.tight_layout()
plt.show()
```

This visualization emphasizes that:

  •	Acute conditions (Pain, Colds & Flu) have the highest number of drugs, indicating frequent treatment needs.

  •	Chronic and lifestyle-related conditions (Hypertension, Diabetes) also make up a significant share, showing the impact of modern health trends.
